<h3> <center> Guass Elimination </center> </h3>

```python
c = np.array([[25,5,1,106.8],[64,8,1,177.2],[144,12,1,279.2]])
x = np.ones(3)
n = len(c)
for i in range(0,n):
    for j in range((i+1),n):
        c[j] = c[j]-(c[i]*(c[j][i])/(c[i][i]))
        

for i in range(n-1,-1,-1):
    y = 0
    for j in range(n-1,-1,-1):
        if i != j:
            if i < j:
                y = y + ((c[i][j]) * x[j])
            else:
                y = y + c[i][j]
    y = c[i][3] - y
    x[i] = y / c[i][i]

    
for i in range(0,int(len(x))):
    print(f"x{i+1} is :", round(x[i],5))
```

<h3> <center> Guass Jaccobi </center> </h3>

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(columns=['x1','x2','x3'])
c = np.array([[2,-1,1,-1],[1,2,-1,6],[1,-1,2,-3]])
x = np.ones(3)

x = np.zeros(len(c))
xcop = np.zeros(len(c))
for k in range(0,15):
    for i in range(0,len(c)):
        y = 0
        for j in range(0,len(c)):
            if i != j:
                y = y + (c[i][j] * x[j])
        y = c[i][len(c)] - y
        xcop[i] = y / c[i][i]
    for i in range(0,len(x)):
        x[i] = xcop[i]
    df.loc[k+1] = [x[0],x[1],x[2]]
        
round(df,4)
```

<h3> <center> Guass Jorden </center> </h3>

<h3> <center> Guass Seidel </center> </h3>

```python
df = pd.DataFrame(columns=['x1','x2','x3','x4'])
c = np.array([[10,-1,2,0,6],[-1,11,-1,3,25],[2,-1,10,-1,-11],[0,3,-1,8,15]])

x = np.zeros(len(c))
for k in range(0,9):
    for i in range(0,len(c)):
        y = 0
        for j in range(0,len(c)):
            if i != j:
                y = y + (c[i][j] * x[j])
        y = c[i][len(c)] - y
        x[i] = y / c[i][i] 
    df.loc[k+1] = [x[0],x[1],x[2],x[3]]
        
round(df,4)
```
